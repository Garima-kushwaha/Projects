```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv(r'Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace = True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.sample(5)
```

####  As Amount values are in float so changing them to int

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

## Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(x='Gender' ,y='Amount' , data=sales_gender)
```

###### *From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.*

### Age

```python
ax = sns.countplot(x='Age Group' , data=df , hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Age_Group = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(x='Age Group' ,y='Amount' , data=sales_Age_Group)
```

###### *From above graphs we can see that most of the buyers are of the Age-Group 26-35 within which female ratio is more than the man.Hence we can say that most of the buyers are female of age-group 26-35.*

# State

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
plt.figure(figsize = (16,5))
sns.barplot(x='State' ,y='Orders' , data=sales_state)
```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
plt.figure(figsize = (16,5))
sns.barplot(x='State' ,y='Amount' , data=sales_state)
```

###### *From the above graphs we can see that most of the Ordērs and Total Sales/Amount are from Uttar Pradesh, Maharashtra, Karnataka.*

# Marital Status

```python
ax = sns.countplot(x='Marital_Status' , data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sns.barplot(x='Marital_Status' ,y='Amount' , data=sales_marital_status, hue='Gender') 
```

###### *From the above Graphs we can see that most of the buyers are Married(Females) with high purchasing power.*

# Occupation

```python
plt.figure(figsize = (20,5))
ax = sns.countplot(x='Occupation' , data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize = (20,5))
sales_occupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x='Occupation' ,y='Amount' , data=sales_occupation) 
```

###### *From the above graphs we can see that most of ther buyers are working in IT Sector, Healthcare, Aviation Sector.*

# Product Category

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Product_Category' , data= df)
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize = (10,5))
sales_product_category = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sns.barplot(x='Product_Category' ,y='Amount' , data=sales_product_category)
plt.xticks(rotation = 90)
```

###### *From the above graphs we can see that most of the sold products are Food, Clothes  & Electronic Gadgets.*

### Top 10 Most Selling Products

```python
plt.figure(figsize = (12,7))
df.groupby('Product_ID')['Orders'].sum().sort_values(ascending = False).head(10).plot(kind='bar')
```
